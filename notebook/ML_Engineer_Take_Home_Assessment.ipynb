```python
from bs4 import BeautifulSoup

def read_xml_file(path):
  with open(path, 'r') as f:
      data = f.read()
  bs_data = BeautifulSoup(data, 'xml')
  return {
      "cat_or_dog": bs_data.find("name").text,
      "xmin": int(bs_data.find("xmin").text),
      "ymin": int(bs_data.find("ymin").text),
      "xmax": int(bs_data.find("xmax").text),
      "ymax": int(bs_data.find("ymax").text),
      "specie": "_".join(path.split(os.sep)[-1].split("_")[:-1])
  }
```

```python
import os
import numpy as np
from PIL import Image
import torch
import torchvision.transforms as transforms
from torch.utils.data import Dataset
from bs4 import BeautifulSoup


class CustomDataset(Dataset):
    def __init__(self, dataset_path, images_list, train=False):
        self.dataset_path = dataset_path
        self.images_list = images_list
        self.train = train

        self.transforms = transforms.Compose([
            transforms.Resize((256, 256)),  # Resize 
            transforms.ToTensor(),  
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # Normalize image
        ])

        image_folder_path = os.path.join(dataset_path, "images")
        label_folder_path = os.path.join(dataset_path, "labels")

        self.data = []
        for image_name in self.images_list:
            image_path = os.path.join(image_folder_path, image_name + ".jpg")
            xml_path = os.path.join(label_folder_path, image_name + ".xml")
            if os.path.exists(xml_path):
                xml_data = read_xml_file(xml_path)
                self.data.append({"image_path": image_path, "xml_data": xml_data})

    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
      data = self.data[index]
      image_path = data["image_path"]
      xml_data = data["xml_data"]

      image = Image.open(image_path).convert("RGB")

      if self.transforms is not None:
          image = self.transforms(image)

      if xml_data.get("xmin") is not None and xml_data.get("ymin") is not None and xml_data.get("xmax") is not None and xml_data.get("ymax") is not None:
          object_label = 1
          cat_or_dog = xml_data["cat_or_dog"]
          xmin = xml_data["xmin"]
          ymin = xml_data["ymin"]
          xmax = xml_data["xmax"]
          ymax = xml_data["ymax"]
          specie = xml_data["specie"]

      else:
          object_label = 0
          cat_or_dog = "FALSE"
          xmin = 0
          ymin = 0
          xmax = 0
          ymax = 0
          specie = "NA"
          
      return image, {
          "cat_or_dog": cat_or_dog,
          "object": object_label,
          "xmin": xmin,
          "ymin": ymin,
          "xmax": xmax,
          "ymax": ymax,
          "specie": specie
      }
```

### Model Class

```python
import torch
import torch.nn as nn
import torchvision.models as models
import torchvision.models.detection as detection_models
```

```python
import torchvision.models as models
import torch.nn as nn

class Model(nn.Module):
    def __init__(self, num_classes):
        super(Model, self).__init__()
        self.backbone = models.resnet18(pretrained=True)
        self.backbone = nn.Sequential(*list(self.backbone.children())[:-2])

        self.avgpool = nn.AdaptiveAvgPool2d(1)
        self.fc = nn.Linear(512, num_classes)

        self.bbox_pred = nn.Linear(512, 4)
        self.object_pred = nn.Linear(512, 1)
        self.cat_or_dog_pred = nn.Linear(512, 2)
        self.specie_pred = nn.Linear(512, 9)

    def forward(self, input):
        features = self.backbone(input)
        features = self.avgpool(features)
        features = features.view(features.size(0), -1)

        object_output = torch.sigmoid(self.object_pred(features))
        cat_or_dog_output = torch.sigmoid(self.cat_or_dog_pred(features))
        specie_output = torch.sigmoid(self.specie_pred(features))

        bbox_output = self.bbox_pred(features)
        xmin_output, ymin_output, xmax_output, ymax_output = torch.split(bbox_output, 1, dim=1)
        xmin_output = xmin_output.squeeze(1)
        ymin_output = ymin_output.squeeze(1)
        xmax_output = xmax_output.squeeze(1)
        ymax_output = ymax_output.squeeze(1)

        return {
            "xmin": xmin_output,
            "ymin": ymin_output,
            "xmax": xmax_output,
            "ymax": ymax_output,
            "object": object_output,
            "cat_or_dog": cat_or_dog_output,
            "specie": specie_output
        }
```

### Training Initialization

```python
from sklearn.preprocessing import LabelEncoder

# Convert string labels to numerical indices
label_encoder = LabelEncoder()
```

```python
import torch

def train(epochs, model_weights=None,resume_training=False):
    model = Model(num_classes=9)
    if model_weights:
        model.load_state_dict(torch.load(model_weights))
        print("Loaded model weights from:", model_weights)
        
    if resume_training:
        model.load_state_dict(torch.load("best_model.pth"))
        print("Resuming training from the best model")

    optimizer = torch.optim.Adam(model.parameters(), lr=0.001)


    # Initialize Loss Functions
    have_object_loss = torch.nn.BCELoss()
    specie_loss = torch.nn.CrossEntropyLoss()
    cat_or_dog_loss = torch.nn.CrossEntropyLoss()
    xmin_loss = torch.nn.MSELoss()
    ymin_loss = torch.nn.MSELoss()
    xmax_loss = torch.nn.MSELoss()
    ymax_loss = torch.nn.MSELoss()

    training_dataset = CustomDataset("/content/assessment_dataset", images_list=train_list)
    training_loader = torch.utils.data.DataLoader(training_dataset, batch_size=32, shuffle=True)

    if torch.cuda.is_available():
        model = model.cuda()

    def train_one_epoch(epoch_index, tb_writer):
        running_loss = 0.
        last_loss = 0.
        for i, data in enumerate(training_loader):
            inputs, labels = data

            numerical_specie_labels = label_encoder.fit_transform(labels["specie"])
            numerical_dorc_labels = label_encoder.fit_transform(labels["cat_or_dog"])

            if torch.cuda.is_available():
                inputs = inputs.cuda()
                labels = {
                    "object": labels["object"].cuda(),
                    "specie": torch.tensor(numerical_specie_labels).cuda(),
                    "cat_or_dog": torch.tensor(numerical_dorc_labels).cuda(),
                    "xmin": labels["xmin"].cuda(),
                    "ymin": labels["ymin"].cuda(),
                    "xmax": labels["xmax"].cuda(),
                    "ymax": labels["ymax"].cuda()
                }
            else:
                labels = {
                    "object": labels["object"],
                    "specie": torch.tensor(numerical_specie_labels),
                    "cat_or_dog": torch.tensor(numerical_dorc_labels),
                    "xmin": labels["xmin"],
                    "ymin": labels["ymin"],
                    "xmax": labels["xmax"],
                    "ymax": labels["ymax"]
                }

            optimizer.zero_grad()
            outputs = model(inputs)

            loss_object = have_object_loss(outputs["object"], labels["object"].unsqueeze(1).float())

            loss_specie = specie_loss(outputs["specie"].float(), labels["specie"].long())
            loss_cat_or_dog = cat_or_dog_loss(outputs["cat_or_dog"].float(), labels["cat_or_dog"].long())

            loss_xmin = xmin_loss(outputs["xmin"].float(), labels["xmin"].float())
            loss_ymin = ymin_loss(outputs["ymin"].float(), labels["ymin"].float())
            loss_xmax = xmax_loss(outputs["xmax"].float(), labels["xmax"].float())
            loss_ymax = ymax_loss(outputs["ymax"].float(), labels["ymax"].float())

            loss = (
                loss_object
                + loss_specie
                + loss_cat_or_dog
                + loss_xmin
                + loss_ymin
                + loss_xmax
                + loss_ymax
            )

            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            if i % 10 == 0:
                last_loss = running_loss / 10
                running_loss = 0.

        return last_loss

    if model_weights:
        model.load_state_dict(torch.load(model_weights))
        print("Loaded model weights from:", model_weights)

    best_loss = float('inf')
    for i in range(epochs):
        epoch_loss = train_one_epoch(i, None)
        print(f'Epoch {i} Loss: {epoch_loss}')

        if epoch_loss < best_loss:
            best_loss = epoch_loss
            torch.save(model.state_dict(), "best_model.pth")
            
    torch.save(model.state_dict(), "final_model.pth")
```

### Visualize

```python
from PIL import Image, ImageDraw
import torchvision.transforms as transforms

def visualize(model_path, image_folder_path, output_folder="output"):
    model = Model(num_classes=9)
    model.load_state_dict(torch.load(model_path))
    model.eval()

    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    image_transforms = transforms.Compose([
        transforms.Resize((256, 256)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    image_details = {}

    for image_name in os.listdir(image_folder_path):
        if image_name.endswith(".jpg") or image_name.endswith(".jpeg"):
            image_path = os.path.join(image_folder_path, image_name)

            try:
                image = Image.open(image_path)
            except:
                continue

            preprocess = image_transforms(image)

            preprocess = preprocess.unsqueeze(0)

            output = model(preprocess)

            # Extract the required details from the output
            has_object = output["object"] > 0.7
            cat_or_dog_labels = output["cat_or_dog"].argmax(dim=1)
            specie_labels = output["specie"].argmax(dim=1)
            xmin = output["xmin"]
            ymin = output["ymin"]
            xmax = output["xmax"]
            ymax = output["ymax"]

            image_details[image_name] = {
                "has_object": has_object.item(),
                "cat_or_dog": "NA" if not has_object else cat_or_dog_labels.item(),
                "specie": "NA" if not has_object else specie_labels.item(),
                "xmin": 0 if not has_object else xmin.item(),
                "ymin": 0 if not has_object else ymin.item(),
                "xmax": 0 if not has_object else xmax.item(),
                "ymax": 0 if not has_object else ymax.item(),
            }

            # Draw bounding box on the image if an object is detected
            if has_object:
                draw = ImageDraw.Draw(image)
                draw.rectangle([(xmin.item(), ymin.item()), (xmax.item(), ymax.item())], outline="red")

            output_path = os.path.join(output_folder, image_name)
            image.save(output_path)

    return image_details
```
